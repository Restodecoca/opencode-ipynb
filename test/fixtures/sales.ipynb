# Sales Analysis

A small notebook that imports `numpy` and `matplotlib`, runs some basic arithmetic, and renders a line plot inline as `image/png`.

```python
import sys
print(f'python {sys.version.split()[0]}')
```

```python
import numpy as np
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
print(f'x.shape={x.shape}, y.shape={y.shape}, y.min={y.min():.3f}, y.max={y.max():.3f}')
```

```python
import io
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from IPython.display import Image, display

fig, ax = plt.subplots(figsize=(4, 2.5))
ax.plot(x, y, label='sin(x)')
ax.set_title('sin(x) on [0, 2pi]')
ax.set_xlabel('x (rad)')
ax.set_ylabel('y')
ax.legend()
ax.grid(True, alpha=0.3)
fig.tight_layout()
buf = io.BytesIO()
fig.savefig(buf, format='png', dpi=100)
plt.close(fig)
display(Image(data=buf.getvalue(), format='png'))
print('plot embedded as image/png')
```

```python
def celsius_to_fahrenheit(c):
    return c * 9 / 5 + 32

samples = [0, 25, 100]
for c in samples:
    print(f'{c} C = {celsius_to_fahrenheit(c):.1f} F')
```
