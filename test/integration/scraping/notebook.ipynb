# Minimal web scrape with the standard library

A minimal scraping example that uses only `urllib.request` and `html.parser`
from the Python standard library. No third-party packages required.

We fetch `https://example.com/`, extract the `<title>`, and print it.
The fetch is wrapped in a `try/except` so that `ipynb_run` returns a
graceful error if the network is down.

```python
import urllib.request
from html.parser import HTMLParser
```

```python
URL = 'https://example.com/'
HEADERS = {'User-Agent': 'opencode-ipynb-example/0.1 (+https://example.com)'}

class TitleParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.title = ''
        self._in_title = False

    def handle_starttag(self, tag, attrs):
        if tag == 'title':
            self._in_title = True

    def handle_data(self, data):
        if self._in_title:
            self.title += data

    def handle_endtag(self, tag):
        if tag == 'title':
            self._in_title = False

try:
    req = urllib.request.Request(URL, headers=HEADERS)
    with urllib.request.urlopen(req, timeout=10) as resp:
        html = resp.read().decode('utf-8', errors='replace')
        status = resp.status
    parser = TitleParser()
    parser.feed(html)
    print(f'status={status}, title={parser.title.strip()!r}')
except Exception as e:
    print(f'fetch failed: {type(e).__name__}: {e}')
```
