# Smoothing a noisy sine wave

A minimal time-series example using only `numpy`.
We generate a sine wave, add Gaussian noise, then smooth it with a simple
moving average (convolution with a uniform window) and report the basic
statistics of the smoothed signal.

```python
import numpy as np

rng = np.random.default_rng(seed=0)
n = 200
t = np.linspace(0.0, 4.0 * np.pi, n)
signal = np.sin(t)
noise = rng.normal(loc=0.0, scale=0.25, size=n)
y_noisy = signal + noise
print(f'n={n}, t.min={t.min():.3f}, t.max={t.max():.3f}')
```

```python
window = 9
kernel = np.ones(window) / window
y_smooth = np.convolve(y_noisy, kernel, mode='same')
print(f'window={window}, y_smooth.shape={y_smooth.shape}')
```

```python
print(f'min={y_smooth.min():.3f}, max={y_smooth.max():.3f}, mean={y_smooth.mean():.3f}')
```
